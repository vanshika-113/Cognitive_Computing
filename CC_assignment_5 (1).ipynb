```python
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')
print("Sum of all elements:",np.sum(gfg))
print("Sum of all elements row-wise:",np.sum(gfg,axis=1))
print("Sum of all elements column-wise:",np.sum(gfg,axis=0))
```

```python
import numpy as np
array=np.array([10,52,62,16,16,54,453])
print("Sorted array:",np.sort(array))
print("Indices of sorted array:",np.argsort(array))
print("4 smallest elements:",np.sort(array)[:4])
print("5 largest elements:",np.sort(array)[-5:])

array=np.array([1.0,1.2,2.2,2.0,3.0,2.0])
print("Integer elements only:",array[array==array.astype(int)])
print("Float elements only:",array[array!=array.astype(int)])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.linspace(-10,10,num=100)
y=x**2
plt.plot(x,y,marker="o")
plt.title("y=x^2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()

y=np.sin(x)
plt.plot(x,y,marker="o")
plt.title("y=sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()

y=np.exp(x)
plt.plot(x,y,marker="o")
plt.title("y=e^x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()

y=np.log(np.abs(x)+1)
plt.plot(x,y,marker="o")
plt.title("y=ln(x+1)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
import numpy as np
sales=np.array([152,202,252,302,352])
tax_rate=((152%5)+5)/100
print("Taxed sales:",sales*(1+tax_rate))
discounted_sales=np.where(sales<152+100,sales*0.95,sales*0.90)
print("Discounted sales:",np.where(sales<152+100,sales*0.95,sales*0.90))
week_sales=np.tile(discounted_sales,(3,1))
print("Three weeks of sales:\n",np.tile(discounted_sales,(3,1)))
weeks_sales_with_increase = week_sales * (1 + 0.02 * np.arange(3)[:, np.newaxis])
print("Three weeks of sales data with 2% increase per week:\n", weeks_sales_with_increase)

```
